```python
import numpy as np
from numpy.linalg import inv
from numpy.linalg import det
import pickle,sys
```

```python
def calculateSquaredLoss(X,y,lambd):
    N = X.shape[0]
    d = X.shape[1]
    w = np.zeros((d,1))
    part_a = ((y-(X.dot(w))).T).dot(y-(X.dot(w)))
    part_b = lambd*((w.T).dot(w))
    squared_loss = (1/2)*(part_a + part_b)
    return squared_loss
```

```python
def learnOLERegression(X,y):
    # Inputs:
    # X = N x d
    # y = N x 1
    # Output:
    # w = d x 1
    w=np.zeros((d,1))
    # IMPLEMENT THIS METHOD
#     return (((y-(X.dot(w))).T).dot((y-(X.dot(w)))))
    #Formula of w taken from the Lectures
    return (inv((X.T).dot(X)).dot(X.T)).dot(y)
```

```python
def testOLERegression(w,Xtest,ytest):
    # Inputs:
    # w = d x 1
    # Xtest = N x d
    # ytest = X x 1
    # Output:
    # mse
    N = Xtest.shape[0]
    d = Xtest.shape[1]
    # IMPLEMENT THIS METHOD
    sum = 0
    for i in range(0,N):
        print("ytest is ",ytest[i])
        print("Xtest is ",Xtest[i])
        sum = sum + (ytest[i]-((w.T).dot(Xtest[i])))**2
    mse = (1/N)*(sum)
    return mse
```
